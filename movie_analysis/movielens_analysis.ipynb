# Movielens电影评分数据分析

```python
import numpy as np
import pandas as pd
```

## 读取数据

```python
# 读取用户信息
users = pd.read_table('./users.dat', header=None, 
                      names=['UserID','Gender','Age','Occupation','Postcode'], 
                      sep='::',
                      engine='python')


# 输出用户总数
print(len(users))
type(users)
```

```python
# 查看前五条记录
users.head(5)
```

```python
# 导入电影评分表
ratings = pd.read_csv('./ratings.dat', header=None, 
                        names=['UserID', 'MovieID', 'Rating', 'TimeStamp'], 
                        sep='::',
                        engine='python')

# 打印总评分条数
print(len(ratings))
type(users)
```

```python
# 查看前五条记录
ratings.head(5)
```

```python
# 导入电影数据表
movies = pd.read_table('./movies.dat', header=None, 
                       names=['MovieID', 'MovieName', 'Category'], 
                       sep='::',
                       engine='python')

#打印总电影数
print(len(movies))
```

```python
# 查看前五条记录
movies.head(5)
```

## 合并数据表

```python
#多张表进行合并  先将users与ratings两张表合并再跟movied合并
data = pd.merge(pd.merge(users, ratings), movies)

#获取尾10行数据
data.tail(10)
```

## 数据初步描述分析

```python
data.describe()
```

```python
data.info()
```

## 查看数据

```python
# 合并后的每一条记录：
# 年龄，职业，性别，邮编，电影ID，评分，时间戳，电影信息，电影分类等一系列信息

data[data.UserID==1].head(10)
```

## 查看电影不同性别平均评分，计算分歧差值，之后排序

```python
gender_data=data.pivot_table(index=['MovieName'],
                             columns=['Gender'],
                             values=['Rating'],
                             aggfunc='mean')

print(gender_data.shape)
gender_data.head(5)
```

```python
# 查看电影分歧最大的那部电影，分歧差值
gender_data['Diff_value']=np.abs(gender_data['Rating']['F']-gender_data['Rating']['M'])
gender_data.head(5)
```

```python
# 男女电影分歧进行排序 
data_gender_sorted=gender_data.sort_values(by=['Diff_value'],
                                           ascending=False)

data_gender_sorted.head(10)
```

## 计算每部电影平均得分并对其进行排序

```python
data_mean_rating=data.pivot_table(index='MovieName',
                                  values='Rating',
                                  aggfunc='mean',
                                  margins=True)

data_mean_rating.head(10)
```

```python
# 对电影平均得分排序
data_mean_rating_sorted=data_mean_rating.sort_values(by='Rating',ascending=False)
data_mean_rating_sorted.head()
```

## 查看评分次数多的电影并进行排序1 

```python
data_rating_num=data.groupby('MovieName').size()

type(data_rating_num)
data_rating_num.head(10)
```

```python
#进行排序
data_rating_num_sorted=data_rating_num.sort_values(ascending=False)
data_rating_num_sorted.head()
```

## 查看评分次数多的电影并进行排序2

```python
movie_comment=pd.crosstab(data.MovieName,data.Rating,margins=True)
movie_comment.head(10)
```

```python
movie_comment_sort=movie_comment.sort_values(by=['All'],ascending=False)
movie_comment_sort.head(10)
```

## 过滤评分条数不足250条的电影

```python
hot_movies=movie_comment_sort[movie_comment_sort['All']>250]

hot_movies.head(20)
```

```python
hot_movies_sorted=hot_movies.sort_values(by=['All'],ascending=False)
hot_movies_sorted[:20]
```

## 评分最高的十部电影


```python
#评分最高的十部电影
movies_stats=data.groupby('MovieName').agg({'Rating':[np.size,np.mean,np.std]})
print(type(movies_stats))
movies_stats.head(10)
```

```python
# 被评论的次数>=100
atleast_100=movies_stats['Rating']['size']>=100
movies_stats[atleast_100].sort_values([('Rating','mean')],ascending=False)[:10]
```

## 在原数据中标记出用户位于的年龄分组

```python
labels=['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79']
data['Age_Group']=pd.cut(data.Age,range(0,81,10),labels=labels)
data.head(5)
```

```python
count_age.plot(kind='bar',color='orange')
mp.xticks(rotation=45)
mp.xlabel('user_age_group')
mp.ylabel('count_User')
mp.show()
```

## 电影评分表中统计不同类型电影的频数

```python
movie_clean_1=pd.DataFrame(movies.Category.str.split('|').tolist(),
                           index=movies.MovieID)

movie_clean_1.head()
```

```python
movie_clean_2=movie_clean_1.stack().reset_index()
movie_clean_2.head()
```

```python
#列重命名为category
movies_category=movie_clean_2.drop('level_1',axis=1).rename(columns={0:'Category'})
movies_category.head()
```

```python
#原电影数据分类替换成单独分类的category
movies=pd.merge(movies.drop('Category',axis=1),movies_category)
movies.head(10)
```

```python
#合并，构建电影评分数据集movie_ratings
movie_ratings=ratings.merge(movies.drop('MovieName',axis=1),how='inner',on='MovieID')
movie_ratings.head()
```

```python
#计算movies_ratings中不同类型电影的频数
movies_ratings_sorted=movie_ratings.groupby(['Category'])['MovieID'].size()
count_movie=movies_ratings_sorted.sort_values(ascending=False)

count_movie.plot(kind='bar',color='green')
mp.xticks(rotation=45)
mp.show()
```
